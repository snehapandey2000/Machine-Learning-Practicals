```python
import requests
from bs4 import BeautifulSoup

url = "https://www.amazon.in/Midnights-Moonstone-Blue-Taylor-Swift/product-reviews/B0BFFB1G2R/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"

page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")
```

```python
reviews_array = []
while True:
    reviews = soup.find_all("div", {"data-hook": "review"})
    for review in reviews:
        reviewer = review.find("span", {"class": "a-profile-name"}).get_text().strip()
        if review.find("a", {"data-hook": "review-title"}):
            review_title = review.find("a", {"data-hook": "review-title"}).get_text().strip()
        else:
            review_title=""
        review_text = review.find("span", {"data-hook": "review-body"}).get_text().strip()
        rating = review.find("span", {"class": "a-icon-alt"}).get_text().strip()

        print("Reviewer:", reviewer)
        print("Review Title:", review_title)
        print("Review Text:", review_text)
        print("Rating:", rating)
        print("\n")
        reviews_array.append({"reviewer": reviewer, "review_title": review_title, "review_text": review_text, "rating": rating})
    # Check if there is a next page
    next_page = soup.find("li", {"class": "a-last"})
    if not next_page:
        break

    # Go to the next page
    url = "https://www.amazon.in" + next_page.find("a").get("href")
    res = requests.get(url)
    soup = BeautifulSoup(res.text, "html.parser")
```

```python
import pandas as pd
# Convert the list of reviews to a pandas DataFrame
df = pd.DataFrame(reviews_array)

# Write the DataFrame to a CSV file
df.to_csv("review_amazon.csv", index=False)
```
